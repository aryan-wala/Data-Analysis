```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("../googleplaystore.csv")
```

```python
df.head()
```

```python
si = SimpleImputer(missing_values = np.nan,strategy='mean')
si.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = si.transform(df.iloc[:,2:3].values)
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

## Name apps that are in ART_AND_DESIGN with rating more than 4.5 in ascending order WRT Raitings?

```python
df_1 = df[df["Category"] == 'FAMILY']
df_1 = df_1[df_1["Rating"] > 4.5]
df_1.sort_values(by = "Rating", ascending=True)
```

## Name apps that are in ART_AND_DESIGN with rating more than 4.5 in descending order WRT Reviews?

```python
df.info()
```

```python
df["Reviews"] = df["Reviews"].astype("float")
```

```python
df_1 = df[df["Category"] == 'FAMILY']
df_1 = df_1[df_1["Rating"] > 4.5]
df_1.sort_values(by = "Reviews", ascending=False)
```
